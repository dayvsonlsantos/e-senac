```python
# Importando pandas
import pandas as pd
```

```python
# Carregar dados do CSV
dados_municipios = pd.read_csv('ATU_MUNICIPIOS_2022.csv', sep=',', low_memory=False)

# ver números de linhas e colunas
dados_municipios.shape
```

```python
#Visualizando as primeiras linhas do Dataframe
dados_municipios.head()
```

```python
# ver tipos das variáveis
dados_municipios.dtypes
```

```python
# Fazendo uma copia da base de dados
dados_municipios_copy = dados_municipios.copy()
```

```python
dados_municipios_copy.head()
```

```python
dados_municipios_copy.dtypes
```

```python
# Apagando colunas desnecessárias p/ o estudo

# Mantendo o código do municípios, pois posteriormente, utilizarei essa coluna, em vez do nome do municipio.

dados_municipios_copy = dados_municipios_copy.drop(columns = ['Ano'])
```

```python
dados_municipios_copy.head()
```

```python
# Base de dados p/ o ensino médio

dados_municipios_ensino_medio = dados_municipios_copy[["Região", "UF", "Código do Município", "Município", "Localização", "Dependência Administrativa", "Total Ensino Médio", "1ª série", "2ª série", "3ª série", "4ª série", "Não-Seriado"]].copy()
dados_municipios_ensino_medio.head()
```

```python
dados_municipios_ensino_medio.describe()
```

```python
import numpy as np
```

```python
# Substituir '--' por NaN

colunas = ["Total Ensino Médio", "1ª série", "2ª série", "3ª série", "4ª série", "Não-Seriado"]

dados_municipios_ensino_medio[colunas] = dados_municipios_ensino_medio[colunas].replace('--', np.nan)
```

```python
dados_municipios_ensino_medio.head()
```

```python
# Substituir , por .

dados_municipios_ensino_medio[colunas] = dados_municipios_ensino_medio[colunas].applymap(lambda x: str(x).replace(',', '.'))
```

```python
# Convertendo para Float:

dados_municipios_ensino_medio[colunas] = dados_municipios_ensino_medio[colunas].astype(float)
dados_municipios_ensino_medio.dtypes
```

```python
# Exibir primeiros dados

dados_municipios_ensino_medio.head()
```

```python
# Convertendo demais colunas em String

columns_to_str = ["Código do Município"]

dados_municipios_ensino_medio[columns_to_str] = dados_municipios_ensino_medio[columns_to_str].astype(str)
dados_municipios_ensino_medio.dtypes
```

```python
# Substituir NaN por 0

colunas = ["Total Ensino Médio", "1ª série", "2ª série", "3ª série", "4ª série", "Não-Seriado"]

dados_municipios_ensino_medio[colunas] = dados_municipios_ensino_medio[colunas].replace(np.nan, 0)
```

```python
dados_municipios_ensino_medio.head()
```

```python
# Criando filtros:

# Urbano e Municipal
filtro_urbano_municipal = (dados_municipios_ensino_medio['Localização'] == 'Urbana') & (dados_municipios_ensino_medio['Dependência Administrativa'] == 'Municipal')
alunos_ensMed_urbano_municipal = dados_municipios_ensino_medio[filtro_urbano_municipal]

# Urbano e Estadual
filtro_urbano_estadual = (dados_municipios_ensino_medio['Localização'] == 'Urbana') & (dados_municipios_ensino_medio['Dependência Administrativa'] == 'Estadual')
alunos_ensMed_urbano_estadual = dados_municipios_ensino_medio[filtro_urbano_estadual]

# Urbano e Federal
filtro_urbano_federal = (dados_municipios_ensino_medio['Localização'] == 'Urbana') & (dados_municipios_ensino_medio['Dependência Administrativa'] == 'Federal')
alunos_ensMed_urbano_federal = dados_municipios_ensino_medio[filtro_urbano_federal]

# Urbano e Privado
filtro_urbano_privado = (dados_municipios_ensino_medio['Localização'] == 'Urbana') & (dados_municipios_ensino_medio['Dependência Administrativa'] == 'Privado')
alunos_ensMed_urbano_privado = dados_municipios_ensino_medio[filtro_urbano_privado]

# Rural e Municipal
filtro_rural_municipal = (dados_municipios_ensino_medio['Localização'] == 'Rural') & (dados_municipios_ensino_medio['Dependência Administrativa'] == 'Municipal')
alunos_ensMed_rural_municipal = dados_municipios_ensino_medio[filtro_rural_municipal]

# Rural e Estadual
filtro_rural_estadual = (dados_municipios_ensino_medio['Localização'] == 'Rural') & (dados_municipios_ensino_medio['Dependência Administrativa'] == 'Estadual')
alunos_ensMed_rural_estadual = dados_municipios_ensino_medio[filtro_rural_estadual]

# Rural e Federal
filtro_rural_federal = (dados_municipios_ensino_medio['Localização'] == 'Rural') & (dados_municipios_ensino_medio['Dependência Administrativa'] == 'Federal')
alunos_ensMed_rural_federal = dados_municipios_ensino_medio[filtro_rural_federal]

# Rural e Privado
filtro_rural_privado = (dados_municipios_ensino_medio['Localização'] == 'Rural') & (dados_municipios_ensino_medio['Dependência Administrativa'] == 'Privado')
alunos_ensMed_rural_privado = dados_municipios_ensino_medio[filtro_rural_privado]
```

```python
# Média de Alunos do Ensino Médio por UF: Área Urbana e escola Municipal
total_alunos_ensMed_urbano_municipal = alunos_ensMed_urbano_municipal[["Região", "UF", "Localização", "Dependência Administrativa", "Total Ensino Médio"]].groupby('UF').sum(numeric_only=True)

total_alunos_ensMed_urbano_municipal.head(20)
```
